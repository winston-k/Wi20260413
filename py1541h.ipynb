```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel('celebs2_corr.xlsx')

# Filter only numerical variables (not excluding 'Sex' which is categorical despite being stored as int64)
numerical_df = df.select_dtypes(include=['float64', 'int64'])
# if 'Sex' in numerical_df.columns:
#    numerical_df = numerical_df.drop('Sex', axis=1)

# Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Create a heatmap for better visualization
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.tight_layout()
plt.show()
```


### Explanation of Correlation Coefficients:

1. **Interpretation Scale**:
   - 1.0: Perfect positive correlation
   - 0.7 to 0.9: Strong positive correlation
   - 0.4 to 0.6: Moderate positive correlation
   - 0.1 to 0.3: Weak positive correlation
   - 0: No correlation
   - -0.1 to -0.3: Weak negative correlation
   - -0.4 to -0.6: Moderate negative correlation
   - -0.7 to -0.9: Strong negative correlation
   - -1.0: Perfect negative correlation

2. **Key Relationships to Look For**:
   - Height and Weight likely have a positive correlation as taller people tend to weigh more
   - Weight_BP might correlate with Weight as higher weight often relates to blood pressure
   - W_noise appears to be Weight with added noise, so it should correlate strongly with Weight
   - The "random" variable should show little to no correlation with other variables

3. **Practical Significance**:
   - Strong correlations (>0.7) suggest variables that move closely together
   - Variables with high correlation might be redundant in predictive models
   - Near-zero correlations indicate independent variables

The heatmap visualization makes it easy to spot patterns and identify which variables have meaningful relationships in your dataset.
