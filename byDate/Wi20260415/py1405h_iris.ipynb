# Practice 2: Iris Dataset Correlation Analysis


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Display the first few rows
print("First 5 rows of the Iris dataset:")
print(iris_df.head())

# Compute correlation matrix
correlation_matrix = iris_df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Iris Dataset Features')
plt.tight_layout()
plt.show()

# Describe the correlations
for i in range(len(iris_df.columns)):
    for j in range(i+1, len(iris_df.columns)):
        col1 = iris_df.columns[i]
        col2 = iris_df.columns[j]
        corr = correlation_matrix.iloc[i, j]
        strength = "strong" if abs(corr) > 0.7 else "moderate" if abs(corr) > 0.3 else "weak"
        direction = "positive" if corr > 0 else "negative"
        print(f"Correlation between {col1} and {col2}: {corr:.3f} ({strength} {direction})")
```


## Explanation of Iris Dataset Correlations:

1. **Petal Length and Petal Width**: These typically show the strongest positive correlation (~0.9), indicating that flowers with longer petals also tend to have wider petals.

2. **Petal Length/Width and Sepal Length**: Usually show moderate to strong positive correlations, suggesting that larger flowers tend to have both larger petals and sepals.

3. **Sepal Width and other features**: Often shows weak or even negative correlations with other features, indicating that sepal width varies somewhat independently from other measurements.

These correlations reflect the morphological relationships in iris flowers and help distinguish between the three iris species in the dataset. The strong correlations between petal measurements are particularly useful for species classification.

# Assistant
The sepal and petal are different parts of a flower with distinct functions and characteristics:

**Sepals:**
- The outermost parts of a flower that form the calyx
- Usually green, leaf-like structures that protect the flower bud before it opens
- Typically provide support and protection for the developing flower
- In iris flowers, sepals are often large and showy (sometimes called "falls")

**Petals:**
- The inner parts of a flower that form the corolla
- Usually colorful and attractive to pollinators
- Primary function is to attract insects and other pollinators
- In iris flowers, petals are typically upright (sometimes called "standards")

In the iris dataset, the measurements of these different flower parts (length and width) help distinguish between the three iris species (setosa, versicolor, and virginica), as each species has characteristic sepal and petal dimensions.
